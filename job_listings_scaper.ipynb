```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://realpython.github.io/fake-jobs/"
all_jobs = []

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
job_cards = soup.find_all("div", class_="card-content")
```

```python
for jobs in job_cards:
    title = jobs.find("h2", class_="title").text.strip()
    company = jobs.find("h3", class_="company").text.strip()
    location = jobs.find("p", class_="location").text.strip()
    job_links = jobs.find("a")
    if job_links:
        job_urls = job_links["href"]
    else:
        job_urls = "NaN"
    
    all_jobs.append({
        "Title": title,
        "Company Name": company,
        "Location": location,
        "Job detail page URL": job_urls
    })
```

```python
df = pd.DataFrame(all_jobs)
df.to_csv("fake_jobs.csv", index=False)
```
